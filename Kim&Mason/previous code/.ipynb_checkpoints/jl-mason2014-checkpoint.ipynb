```julia
using ForwardDiff, Plots, NLsolve
```

Neglect electrostatic.  
Use the parameter values from Mason 2014 paper.  
Take $\phi = 0.6$.

```julia
alpha = 0.74  # Shear effect parameter
sigma = 9.8e-3  # Surface tension in J/mÂ²
k_B = 1.38e-23  # Boltzmann constant in J/K
T = 298  # Temperature in K
a = 530e-9  # Droplet radius in meters
xi = 0.14  # Coupling parameter
phi_c = 0.625  # Critical volume fraction
```

```julia
phi_T_squared = (3 * k_B * T / a^3) / (2 * pi * xi * sigma / a)

function phid_star(phi, gamma)
    term1 = phi - (phi_c - alpha * gamma^2)
    return 0.5 * (term1 + sqrt(term1^2 + phi_T_squared))
end 
```

```julia
phi_values = 0.55:0.0003:0.85
gamma_fixed = 0.006
phid_star_values = phid_star.(phi_values, gamma_fixed)
plot(phi_values, phid_star_values, xlabel="phi", ylabel="Î¦d*", title="Î¦d* vs phi, (Î³ = $gamma_fixed)", lw=2)
```

```julia
gamma_values = 0:0.0001:0.01
phi_fixed =0.6
phid_star_values = phid_star.(phi_fixed, gamma_values)

plot(gamma_values, phid_star_values, xlabel="Gamma (Î³)", ylabel="Î¦d*", title="Ï_d* as a Function of Î³ (Ï = $phi_fixed)", lw=2)
```

```julia
phi_values = range(0.55, 0.85, length=300)
gamma_values = range(0.0, 0.01, length=300)

phid_star_matrix = [phid_star(phi, gamma) for phi in phi_values, gamma in gamma_values]

# Plotting Ï_d* as a 3D surface plot as a function of Ï and Î³
surface(phi_values, gamma_values, phid_star_matrix, xlabel="Ï", ylabel="Î³", zlabel="Ï_d*", title="Ï_d* as a Function of Ï and Î³", c=:viridis)
```

```julia
function F(phi, gamma)
    return 4 * Ï * a^2 * sigma * xi * phid_star(phi,gamma)^2 - 3 * k_B * T * log(phi_c + phid_star(phi,gamma) - phi - alpha * gamma^2)
end
```

```julia
function osmotic_pressure(phi)
    dF_dphi = ForwardDiff.derivative(g -> F(g, 0), phi)
    return (phi^2 / ((4 / 3) * Ï * a^3)) * dF_dphi
end
```

```julia
phi_values = range(0.55, 0.85, length=300)
pi_values = [osmotic_pressure(phi) for phi in phi_values]
```

```julia
plot(phi_values, (pi_values / sigma * a) , xlabel="Ï", ylabel="Î ", title="Osmotic Pressure Î  as a Function of Ï", legend=false, yscale=:log)
```

```julia
function F(phi_d, phi, gamma)
    term = phi_c + phi_d - phi - alpha * gamma^2
    if term <= 0
        return Inf  # Return a large value to avoid invalid log
    end
    return 4 * Ï * a^2 * sigma * xi * phi_d^2 - 3 * k_B * T * log(term)
end

function find_phid_star(phi, gamma)
    f_prime(phi_d_vec) = ForwardDiff.derivative(phi_d -> F(phi_d, phi, gamma), phi_d_vec[1])
    
    result = nlsolve(f_prime, [1e-15]; inplace=false)  # Initial guess without bounds
    phi_d_star = result.zero[1]  

    boundary_condition = phi_c - phi - alpha * gamma^2

    if boundary_condition > 0
        lower_bound = 0
    else
        lower_bound = phi + alpha * gamma^2 - phi_c
    end

    if phi_d_star < lower_bound
        return lower_bound
    end
    return phi_d_star
end

phi_values = range(0.55, 0.85, length=100)
gamma_values = range(0.0, 0.01, length=100)

phid_star_matrix = [find_phid_star(phi, gamma) for phi in phi_values, gamma in gamma_values]

# Plotting Ï_d* as a 3D surface plot as a function of Ï and Î³
surface(phi_values, gamma_values, phid_star_matrix, xlabel="Ï", ylabel="Î³", zlabel="Ï_d*", title="Ï_d* as a Function of Ï and Î³", c=:viridis)

# Display the plot
plot!()
```

```julia
function target_eqn(phi_d, phi, gamma)
    return 4 * Ï * a^2 * sigma * xi * phi_d^2 - 3 * k_B * T * log(phi_c + phi_d + phi - alpha * gamma^2)
end

function find_min_phi_d(phi, gamma)
    min_phi_d = 0.0
    min_value = target_eqn(0.0, phi, gamma)

    for phi_d in range(0, 0.35, length=100)
        current_value = target_eqn(phi_d, phi, gamma)
        if current_value < min_value
            min_value = current_value
            min_phi_d = phi_d
        end
    end
    
    return min_phi_d, min_value
end

phi_values = range(0.55, 0.85, length=50)
gamma_values = range(0, 0.01, length=50)

phi_d_values = zeros(length(phi_values), length(gamma_values))

for i in 1:length(phi_values)
    for j in 1:length(gamma_values)
        phi = phi_values[i]
        gamma = gamma_values[j]
        phi_d, _ = find_min_phi_d(phi, gamma)  
        phi_d_values[i, j] = phi_d
    end
end

phi_grid = repeat(phi_values', 1, length(gamma_values))
gamma_grid = repeat(gamma_values, 1, length(phi_values))
# surface(phi_grid, gamma_grid, phi_d_values', xlabel="phi", ylabel="gamma", zlabel="phi_d", title="3D Surface of phi_d")
heatmat(log.(phi_d_values)', xlabel="Ï", ylabel="Î³", zlabel="Ï_d", title="Ï_d as a Function of Ï and Î³")
```
